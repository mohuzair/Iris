```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
iris = load_iris()
irisData = iris.data
colNam = iris.feature_names
df = pd.DataFrame(irisData,columns=colNam)
df.head(10)
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,10))
df['sepal length (cm)'].hist(ax=axes[0][0])
df['sepal width (cm)'].hist(ax=axes[0][1])
df['petal length (cm)'].hist(ax=axes[1][0])
df['petal width (cm)'].hist(ax=axes[1][1])
plt.show()
```

```python
df.corr()
```

```python
X = df
Y = iris.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=42)
```

```python
X_test
```

```python
#Linear Regression

model = LinearRegression()
model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)
sc_lr = round(model.score(X_test, Y_test) * 100 , 2)

print("Accuracy: ", str(sc_lr) ,"%" )
```

```python
#Logistic Regression

model2 = LogisticRegression()
model2.fit(X_train,Y_train)

Y_pred = model2.predict(X_test)
sc_logr = round(model2.score(X_test, Y_test) * 100,2)

print("Accuracy: ", str(sc_logr) ,  " %")
```
